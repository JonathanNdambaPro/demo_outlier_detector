```
from mdo_outlier_detector.detector import Mdo
```

```
from google.colab import drive
import os
import pandas as pd
import plotly.express as px
from typing import List

drive.mount('/content/drive')

os.chdir('/content/drive/My Drive/MDO_data')

df_SnP500 = pd.read_csv("^GSPC.csv")
df_VIX = pd.read_csv("^VIX.csv")
df_stock = pd.read_csv("data_stocks.csv")
```

```
test = Mdo()

Open = 'Open'
High = 'High'
Low = 'Low'
Close = 'Close'


params = {
    'inference_type': "bayesian",
    'n_components': 10

}

def test_MDO(df: pd.DataFrame, Var_list: List, **params) -> pd.DataFrame:
  X = df[Var_list]
  X_scoring = X.copy()
  X_scoring = test.transform(X, **params)

  return X_scoring

List_Var = [f'{Open}', f'{High}', f'{Low}']

X_scoring = test_MDO(df_SnP500, List_Var, **params)

fig = px.scatter_3d(X_scoring, x=f'{Open}', y=f'{High}', z=f'{Low}',
                    color='global_metrics')
fig.show()

```

```
List_Var = [f'{Open}', f'{Close}']

X_scoring = test_MDO(df_SnP500, List_Var, **params)

fig = px.scatter(X_scoring, x=f"{Open}", y=f"{Close}", color='global_metrics')
fig.show()
```
